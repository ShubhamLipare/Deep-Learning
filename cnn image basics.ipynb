### cv2 by default read it as BGR

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

### to convert it into RGB we can use cvtcolor function

```python
img2=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```

### Gray scale

```python
img3 = cv2.imread(path,cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img3,cmap="gray")
```
